```python
import pandas as pd
from catboost import CatBoostRegressor, Pool
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df_train = pd.read_csv('train_dataset_train.csv', sep=';', index_col=None, 
                    dtype={'PATIENT_SEX':str, 'MKB_CODE':str, 'ADRES':str,
                           'VISIT_MONTH_YEAR':str, 'AGE_CATEGORY':str, 'PATIENT_ID_COUNT':int})
df_test = pd.read_csv('test_dataset_test.csv', sep=';', index_col=None, 
                   dtype={'PATIENT_SEX':str, 'MKB_CODE':str, 'ADRES':str, 'VISIT_MONTH_YEAR':str, 'AGE_CATEGORY':str})
```

```python
train = df_train.copy()
test = df_test.copy()
```

```python
# выделяем общий класс - портрет пациента (PATIENT_SEX+ADRES+AGE_CATEGORY)
def id_patient (data):
    data['ID_PATIENT'] = data['PATIENT_SEX'] + ':' + data['AGE_CATEGORY'] + ':' + data['MKB_CODE'] + ':' + data['ADRES'] 
    #data = data.drop(columns=['PATIENT_SEX', 'AGE_CATEGORY', 'MKB_CODE', 'ADRES']) 
    return data
```

```python
train = id_patient (train)
test = id_patient (test)
```

```python
train.head()
```

```python
#избавимся от ID которых нет в тесте         !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
def min_df (train, test):
    id_patient_test = test['ID_PATIENT'].unique()
    abc = train[train['ID_PATIENT'].isin(id_patient_test)].reset_index(drop=True)
    return abc
#def abcd (data):
#    visit = pd.DataFrame(data['ID_PATIENT'].astype(str))
#    visit = visit['ID_PATIENT'].str.split(':',expand=True)
#    visit.columns = ['1','2', '3', '4']
#    visit = pd.concat([data, visit],axis=1)
##    visit = visit.drop(columns=['ID_PATIENT'])
##    visit = visit.drop(columns=['VISIT_YEAR']) 
#    return visit
#train = abcd (train)
#test_test = abcd (test_test)
#test = abcd (test)
#train.head(20)
```

```python
train.head(5)
```

```python
#делаем промежуточную тестовую выборку
test_test = train [train['VISIT_MONTH_YEAR'] == '03.22'].reset_index(drop=True)
train = train.loc[train['VISIT_MONTH_YEAR'] != '03.22'].reset_index(drop=True)
```

```python
train = min_df (train, test_test)
```

```python
train['COUNT'] = train.groupby(['ID_PATIENT'])['ID_PATIENT'].transform('count')
```

```python
train['SUM'] = train.groupby(['COUNT'])['COUNT'].transform('count') #, 'ADRES', 'VISIT_YEAR'
train['K'] = (train['SUM'].max()/train['SUM']).astype(int)
```

```python
import numpy as np
koef = train['K'].unique()
train_big = pd.DataFrame() 
for i in koef:
    df_koef = train[train['K'] == i]
    df_koef = pd.DataFrame(np.repeat(df_koef.values, i, axis=0), columns=df_koef.columns).astype(df_koef.dtypes)
    train_big = pd.concat([train_big, df_koef],axis=0)
train = train_big
```

```python
train = train.drop(columns=['COUNT', 'SUM', 'K'])
```

```python
train.head()
```

```python
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
train = train.drop(columns=['ID_PATIENT'])
test_test = test_test.drop(columns=['ID_PATIENT'])
test = test.drop(columns=['ID_PATIENT'])
```

```python
X = train.drop(columns=['PATIENT_ID_COUNT'])
y = train['PATIENT_ID_COUNT']
X_test = test_test.drop(columns=['PATIENT_ID_COUNT'])
y_test = test_test['PATIENT_ID_COUNT']
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.20, random_state=12345)
```

```python
#column_category = ['VISIT_MONTH_YEAR', 'ID_PATIENT'] 
column_category = ['PATIENT_SEX', 'AGE_CATEGORY', 'MKB_CODE', 'ADRES', 'VISIT_MONTH_YEAR']# 'ID_PATIENT'] 
pool_train = Pool(X_train, y_train, cat_features = column_category)
pool_valid = Pool(X_valid, cat_features = column_category)
pool_test = Pool(X_test, cat_features = column_category)
pool_test_solution = Pool(test, cat_features = column_category)
```

```python
#Объявление CatBoostRegressor и обучение
model = CatBoostRegressor(
    #task_type='GPU',
    random_state = 12345,
    iterations = 100,
    learning_rate = 0.2,
    depth = 16,
    eval_metric='R2',
    bootstrap_type ='Bernoulli',
    verbose = 10
    )
model.fit(pool_train, early_stopping_rounds=20, eval_set=(X_valid, y_valid))
```

```python
pred_valid = model.predict(pool_valid)
```

```python
pred_test = model.predict(pool_test)
```

```python
print("Значение метрики R2 на valid: ", r2_score(y_valid, pred_valid))
print("Значение метрики R2 на test: ", r2_score(y_test, pred_test))
```
